```python

rating_order = netflix_df_copy['rating'].value_counts().index

# Create the plot
ax = sns.countplot(data=netflix_df_copy, x='rating', order=rating_order, palette='Set2')
ax.set_xticklabels(ax.get_xticklabels(), rotation=90, ha="right")

# Set figure size and title
plt.gcf().set_size_inches(13, 13)
plt.title('Rating')
plt.show()
```

```python
#Relation between Type and Rating of Movies and Shows
plt.figure(figsize=(10,8))
sns.countplot(x='rating',hue='type',data=netflix_df_copy)
plt.title('Relation between Type and Rating of Movies and Shows Seperately')
plt.show();
```

```python
#Relation between Movies and TV Shows
ax = sns.countplot(data=netflix_df_copy, x='type', palette=['#1f77b4', '#ff7f0e'])
plt.gcf().set_size_inches(10, 10)
plt.title('Type')
plt.show()
```

```python
# Which country has the most number of titles produced?
most_titles_produced = netflix_df_copy.groupby('country').count().sort_values('title',ascending=False).head(5)

most_titles_produced.reset_index(inplace=True)
most_titles_produced
```

```python
plt.bar(most_titles_produced.country,most_titles_produced.title)
```

```python
# What are the top 5 most popular ratings on Netflix?
netflix_df_copy_rating = netflix_df_copy['rating'].value_counts()
netflix_df_copy_rating = pd.DataFrame(netflix_df_copy_rating).reset_index()
netflix_df_copy_rating.columns = ['rating', 'Nbr']


colors = ['#e41a1c', '#377eb8', '#4daf4a', '#984ea3', '#ff7f00']
sns.barplot(x='rating', y='Nbr', data=netflix_df_copy_rating.head(5), palette=colors)


plt.title('Top 5 Ratings by Count')
plt.show()
```

```python
# Age-Appropriate Content Distribution
def rating_category(rating):
    if rating in ['G', 'TV-G', 'TV-Y', 'TV-Y7']:
        return 'Kids'
    elif rating in ['PG', 'TV-PG', 'TV-Y7-FV']:
        return 'Teens'
    elif rating in ['PG-13', 'TV-14']:
        return 'Young Adults'
    elif rating in ['R', 'TV-MA', 'NC-17']:
        return 'Adults'
    else:
        return 'Unknown'

netflix_df_copy['age_group'] = netflix_df_copy['rating'].apply(rating_category)
sns.countplot(data=netflix_df_copy, x='age_group', order=['Kids', 'Teens', 'Young Adults', 'Adults', 'Unknown'])
plt.title('Content Distribution by Age Group')
```

```python
#Correlate Rating with Duration (for Movies)

movie_df = netflix_df_copy[netflix_df_copy['type'] == 'Movie'].copy()
movie_df['duration_min'] = movie_df['duration'].str.extract('(\d+)').astype(float)

rating_order = movie_df.groupby('rating')['duration_min'].median().sort_values().index
plt.figure(figsize=(12, 6))
sns.boxplot(data=movie_df, x='rating', y='duration_min', order=rating_order)
plt.xticks(rotation=90)
plt.ylabel('Duration (Minutes)')
plt.title('Movie Duration by Rating')
plt.tight_layout()
plt.show()
```

```python
# Ratings by Country
country_rating = netflix_df_copy.groupby(['country', 'rating']).size().unstack().fillna(0)
country_rating = country_rating[country_rating.sum(axis=1) > 20]  # filter for countries with enough data
country_rating.plot(kind='bar', stacked=True, figsize=(15, 7))
plt.title('Rating Distribution by Country')
```

```python
#Top Genres by Rating

df_genres = netflix_df_copy.copy()
df_genres['genre'] = df_genres['listed_in'].str.split(', ')
df_genres = df_genres.explode('genre')

top_genres_by_rating = df_genres.groupby(['rating', 'genre']).size().reset_index(name='count')
top5 = top_genres_by_rating.groupby('rating').apply(lambda x: x.nlargest(3, 'count')).reset_index(drop=True)

sns.catplot(data=top5, x='genre', y='count', hue='rating', kind='bar', height=6, aspect=2)
plt.xticks(rotation=45)
plt.title('Top Genres by Rating')
```

```python
# Content Diversity by Country
genre_country_df = netflix_df_copy[['country', 'listed_in']].dropna()
genre_country_df['genre'] = genre_country_df['listed_in'].str.split(', ')
genre_country_df = genre_country_df.explode('genre')

diversity = genre_country_df.groupby('country')['genre'].nunique().sort_values(ascending=False).head(10)
diversity.plot(kind='barh', figsize=(10,6), title='Top 10 Countries by Genre Diversity on Netflix')
```
